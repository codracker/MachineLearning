- Is an unsupervised iterative clustering technique
- It partitions the given data set into k predefined distinct clusters
- A cluster is defined as a collection of data points

```python
import pandas as pd #for reading and writing spreadsheets
import numpy as np #for carrying out efficient computations
import matplotlib.pyplot as plt  #for visualization of data
from sklearn.cluster import KMeans 
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline
```

```python
df = pd.read_csv('income.csv')
df
```

```python
df.shape
```

```python
df.info()
```

```python
plt.scatter(df['Age'],df['Income($)'])
plt.xlabel('Age')
plt.ylabel('Income($)')
```

```python
km = KMeans(n_clusters = 3)
y_pred = km.fit_predict(df[['Age','Income($)']])
y_pred
```

```python
df['Cluster'] = y_pred
df
```

```python
km.cluster_centers_
```

```python
df1 = df[df.Cluster == 0]
df2 = df[df.Cluster == 1]
df3 = df[df.Cluster == 2]
```

```python
df1
```

```python
plt.scatter(df1['Age'],df1['Income($)'],color = 'g')
plt.scatter(df2['Age'],df2['Income($)'],color = 'r')
plt.scatter(df3['Age'],df3['Income($)'],color = 'y')
```

```python
plt.scatter(df1['Age'],df1['Income($)'],color = 'g')
plt.scatter(df2['Age'],df2['Income($)'],color = 'r')
plt.scatter(df3['Age'],df3['Income($)'],color = 'y')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('Age')
plt.ylabel('Income ($)')
plt.legend()
```

```python
#Scaling
scaler = MinMaxScaler()
scaler.fit(df[['Income($)']])
df['Income($)'] = scaler.transform(df[['Income($)']])


scaler.fit(df[['Age']])
df['Age'] = scaler.transform(df[['Age']])
```

```python
df['Income($)'] 
```

```python
df['Age']
```

```python
plt.scatter(df['Age'],df['Income($)'])
plt.xlabel('Age')
plt.ylabel('Income($)')
```

```python
km = KMeans(n_clusters = 3)
y_pred = km.fit_predict(df[['Age','Income($)']])
y_pred
```

```python
df['Cluster'] = y_pred
df
```

```python
km.cluster_centers_
```

```python
df1 = df[df.Cluster == 0]
df2 = df[df.Cluster == 1]
df3 = df[df.Cluster == 2]
```

```python
plt.scatter(df1['Age'],df1['Income($)'],color = 'g')
plt.scatter(df2['Age'],df2['Income($)'],color = 'r')
plt.scatter(df3['Age'],df3['Income($)'],color = 'y')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('Age')
plt.ylabel('Income ($)')
plt.legend()
```

```python
#Elbow
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters = k)
    km.fit(df[['Age','Income($)']])
    sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of Squared Error')
plt.plot(k_rng,sse)
```
